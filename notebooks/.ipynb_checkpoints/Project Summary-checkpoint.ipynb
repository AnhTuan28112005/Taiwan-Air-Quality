# Project Summary

## Key Findings (Các phát hiện chính & Giá trị thực tiễn)

Sau quá trình phân tích dữ liệu và xây dựng mô hình, nhóm nghiên cứu đã rút ra 4 kết luận mang tính chiến lược, có thể chuyển đổi thành hành động thực tế (Actionable Insights) như sau:

### 1. PM2.5 là "Chỉ số vàng" trong giám sát chất lượng không khí
* **Phát hiện:** Kết quả phân tích tương quan cho thấy **PM2.5** có độ tương quan mạnh nhất (**Pearson $\approx$ 0.95**) với chỉ số AQI tổng thể, đóng vai trò là biến đại diện (proxy) hoàn hảo nhất. Ngược lại, các khí như NO hay SO2 có mức độ ảnh hưởng thấp hơn nhiều.
* **Giá trị thực tiễn (Đầu tư):** Đề xuất chiến lược **"Đo diện rộng thay vì đo chuyên sâu"**. Với ngân sách hạn hẹp, địa phương nên ưu tiên lắp đặt mạng lưới cảm biến PM2.5 giá rẻ mật độ cao thay vì đầu tư ít trạm quan trắc đắt tiền đo đủ 6 chất. Việc này giúp nắm bắt 95% bức tranh rủi ro sức khỏe với chi phí tối ưu.

### 2. "Lỗ hổng" phía Tây Bắc và Ngưỡng làm sạch tự nhiên
* **Phát hiện:** Gió không làm sạch không khí một cách tuyến tính. Phải đạt đến **ngưỡng tốc độ 3.8 - 4.3 m/s** thì hiệu quả khuếch tán bụi mới rõ rệt. Đáng chú ý, hướng gió từ **Tây - Tây Bắc - Bắc** luôn mang theo ô nhiễm nặng (Median AQI cao), trong khi gió Nam mang lại không khí sạch.
* **Giá trị thực tiễn (Quy hoạch):** Dữ liệu chỉ ra nguồn phát thải chính nằm ở phía Tây/Bắc các trạm quan trắc. Kiến nghị tuyệt đối **không quy hoạch** khu dân cư mới, trường học hay bệnh viện ở vị trí xuôi gió (phía Đông/Nam) của các cụm công nghiệp hiện hữu ở phía Bắc.

### 3. Rủi ro mang tính "Mùa vụ" cực đoan
* **Phát hiện:** Ô nhiễm tại Đài Loan tuân theo chu kỳ khắc nghiệt. **Mùa Đông (Tháng 10 - Tháng 3)** là "mùa chết" với các đợt ô nhiễm kéo dài trung bình **75 giờ/đợt** (cá biệt >300 giờ). Ngược lại, Mùa Hè gần như sạch hoàn toàn.
* **Giá trị thực tiễn (Y tế):** Cần phân bổ nguồn lực y tế tập trung vào Quý 1 và Quý 4. Các chiến dịch truyền thông về đeo khẩu trang và bảo vệ hô hấp cần được kích hoạt mạnh mẽ từ tháng 10 hàng năm thay vì dàn trải đều quanh năm.

### 4. Khả năng "Nhìn trước tương lai" với độ trễ 6 giờ
* **Phát hiện:** Mô hình Machine Learning (XGBoost) đạt độ chính xác ấn tượng (MAE ~3.88, RMSE ~5.55) cho dự báo **1 giờ tới** và duy trì độ tin cậy tốt trong phạm vi **6 giờ**. Độ chính xác giảm dần sau mốc 12-24 giờ.
* **Giá trị thực tiễn (Công nghệ):** Kết quả này đủ tiêu chuẩn để xây dựng **Hệ thống Cảnh báo Sớm (Early Warning System)**. Thay vì chỉ báo cáo chỉ số hiện tại, hệ thống có thể cảnh báo người dân trước 6 tiếng về các đợt bụi tăng vọt, giúp các trường học và nhóm nhạy cảm chủ động phòng tránh.

---

## Limitations (Hạn chế của Dự án)

Mặc dù đạt được những kết quả khả quan, dự án vẫn tồn tại một số hạn chế cần nhìn nhận:

1.  **Thiên kiến dữ liệu theo không gian (Spatial Bias):** Dữ liệu tập trung quá nhiều vào các trạm lớn (như Chaozhou) trong khi một số huyện/thành phố khác có lượng bản ghi ít hơn. Điều này có thể khiến mô hình dự báo rất tốt ở các vùng trọng điểm nhưng kém chính xác ở vùng sâu vùng xa.
2.  **Khả năng dự báo cực trị (Extreme Events):** Mô hình hiện tại hoạt động tốt ở dải nồng độ trung bình, nhưng có xu hướng dự báo thấp hơn thực tế (under-estimation) đối với các sự kiện ô nhiễm kỷ lục (Outliers > 100 $\mu g/m^3$) do số lượng mẫu học được từ các sự kiện này còn ít.
3.  **Giới hạn khung thời gian dự báo:** Độ chính xác của mô hình giảm nhanh sau 12 giờ. Điều này hạn chế khả năng ứng dụng cho việc lập kế hoạch dài hạn (ví dụ: dự báo cho ngày hôm sau).

---

## Future Directions (Hướng phát triển)

Dựa trên các kết quả và hạn chế trên, nhóm đề xuất các hướng phát triển tiếp theo:

1.  **Nâng cấp mô hình với Deep Learning:** Thử nghiệm các kiến trúc mạng nơ-ron hồi quy như **LSTM** hoặc **GRU** để nắm bắt tốt hơn các phụ thuộc dài hạn, nhằm cải thiện độ chính xác cho dự báo 24 giờ - 48 giờ.
2.  **Tích hợp dữ liệu ngoài (External Data):** Bổ sung thêm các nguồn dữ liệu về **lưu lượng giao thông** và **hoạt động sản xuất công nghiệp** theo thời gian thực để giải thích rõ hơn nguyên nhân của các đợt ô nhiễm từ hướng Tây Bắc, thay vì chỉ dựa vào khí tượng.
3.  **Triển khai ứng dụng thực tế (Deployment):** Xây dựng một Dashboard tương tác (sử dụng Streamlit) tích hợp mô hình XGBoost, cho phép người dân nhập vị trí và nhận khuyến nghị sức khỏe cụ thể trong 6 giờ tới.

- Tôi cũng gặp phải lỗi `DtypeWarning` / các cột có kiểu dữ liệu hỗn hợp (ví dụ: các cột số chứa các ký hiệu đặc biệt như `-`), điều này sau đó gây ra lỗi trong biểu đồ phân tán/tương quan (không thể chuyển đổi chuỗi thành số thực).

- **Cách tôi khắc phục:** Tôi đã sử dụng các bước tải/làm sạch an toàn hơn như `low_memory=False` và chuyển đổi các cột cần thiết bằng `pd.to_numeric(..., errors="coerce")`, sau đó chỉ áp dụng `dropna()` trên các cột cần thiết để tránh mất quá nhiều dữ liệu.

**Những trở ngại phân tích (lựa chọn các chất ô nhiễm “tốt nhất” cho việc giám sát hạn chế — Câu 3)**
- Thách thức chính là chỉ “tương quan cao” thôi là chưa đủ: một chất ô nhiễm có thể tương quan tốt với chỉ số chất lượng không khí (AQI) nói chung nhưng lại không ổn định giữa các địa điểm hoặc bị ảnh hưởng mạnh bởi các điều kiện cục bộ.
- **Cách tôi khắc phục:** Tôi đã kết hợp nhiều tiêu chí:

    - Hệ số tương quan Pearson/Spearman với AQI (mối quan hệ tổng thể),

    - Sự khác biệt trung vị giữa các mức độ trạng thái AQI (khả năng diễn giải rủi ro),

    - Tính nhất quán giữa các địa điểm (độ bền vững giữa các trạm).

Điều này giúp tránh việc chọn một biến chỉ có kết quả tốt ở một tập hợp con các trạm.

**Những trở ngại về mặt khái niệm (ảnh hưởng của gió — Câu hỏi 4)**
- Đối với phân tích gió, mối quan hệ không chỉ đơn giản là tuyến tính: AQI có thể thay đổi khác nhau ở gió nhẹ so với gió mạnh, và hướng gió tạo thêm một lớp phức tạp.

- Phần khó nhất là kiểm tra xem “Liệu có ngưỡng tốc độ gió nào mà AQI giảm rõ rệt hay không?” mà không đưa ra kết luận quá mức.

- **Cách tôi đã vượt qua:** Tôi đã sử dụng:

    - Mật độ Hexbin + xu hướng trung vị theo hướng gió (để thấy được mô hình tổng thể một cách mạnh mẽ),

    - Phân tích theo hướng gió (biểu đồ cực cho chỉ số AQI trung vị theo hướng gió),

    - Quét ngưỡng (delta_median và delta_p_bad so với ngưỡng t) để phát hiện xem sự phân tách có ý nghĩa ở một phạm vi tốc độ gió nào đó hay không.

**Điều gì là thách thức nhất và tại sao?**
- Phần thách thức nhất là đảm bảo phân tích vẫn **đáng tin cậy** bất chấp dữ liệu thực tế phức tạp (giá trị thiếu, các loại dữ liệu hỗn hợp và mối quan hệ phi tuyến tính).

- Một thách thức khác là trình bày kết quả theo cách vừa **dựa trên dữ liệu** vừa **dễ hiểu**, đặc biệt là đối với câu hỏi 3 (khuyến nghị giống như chính sách) và câu hỏi 4 (cơ chế môi trường).

**Kiến thức chuyên môn thu được**:
Hiểu rõ hơn rằng:

- Các biến số liên quan đến bụi mịn thường phản ánh mạnh mẽ rủi ro chất lượng không khí (AQI),

- Gió có thể làm loãng chất ô nhiễm, nhưng hướng gió và địa lý địa phương có thể tạo ra những ngoại lệ,

- Kết luận về “không khí sạch” phải dựa trên bằng chứng (xu hướng + sự khác biệt + tính nhất quán), chứ không phải trực giác.

**Điều gì làm tôi ngạc nhiên nhất?**
- Tôi ngạc nhiên khi một số biểu đồ bị lỗi không phải vì logic sai, mà vì tập dữ liệu chứa các ký tự giữ chỗ ẩn (ví dụ: `-`) bên trong các cột số.

- Tôi cũng thấy thú vị khi tác động của gió không đồng nhất: hiệu ứng có thể phụ thuộc vào phạm vi (gió nhẹ so với gió vừa) và các khu vực hướng gió.

**Dự án này đã định hình sự hiểu biết của tôi về khoa học dữ liệu như thế nào?**
- Dự án này đã giúp tôi nhận ra rằng khoa học dữ liệu không chỉ là việc lập mô hình — mà là một quy trình làm việc hoàn chỉnh:

**chất lượng dữ liệu → làm sạch → khám phá → trực quan hóa cẩn thận → phân tích mạnh mẽ → diễn giải thực tiễn.**
- Nó cũng củng cố tầm quan trọng của việc viết các phân tích có thể hỗ trợ các quyết định thực tế:

- Câu hỏi 3 yêu cầu đưa ra khuyến nghị trong điều kiện ràng buộc (số lượng cảm biến hạn chế),

- Câu hỏi 4 yêu cầu lập luận dựa trên bằng chứng về một cơ chế thực tế (gió và chất lượng không khí).

- Tính ổn định giữa các địa điểm khác nhau.

- Đối với phân tích môi trường, tôi đã học được rằng việc kiểm tra **các mẫu phi tuyến tính** và **hiệu ứng ngưỡng** thường có ý nghĩa hơn là áp đặt các giả định tuyến tính.

**Kiến thức chuyên môn thu được**:
Hiểu rõ hơn rằng:

- Các biến số liên quan đến bụi mịn thường phản ánh mạnh mẽ rủi ro chất lượng không khí (AQI),

- Gió có thể làm loãng chất ô nhiễm, nhưng hướng gió và địa lý địa phương có thể tạo ra những ngoại lệ,

- Kết luận về “không khí sạch” phải dựa trên bằng chứng (xu hướng + sự khác biệt + tính nhất quán), chứ không phải trực giác.

**Điều gì làm tôi ngạc nhiên nhất?**
- Tôi ngạc nhiên khi một số biểu đồ bị lỗi không phải vì logic sai, mà vì tập dữ liệu chứa các ký tự giữ chỗ ẩn (ví dụ: `-`) bên trong các cột số.

- Tôi cũng thấy thú vị khi tác động của gió không đồng nhất: hiệu ứng có thể phụ thuộc vào phạm vi (gió nhẹ so với gió vừa) và các khu vực hướng gió.

**Dự án này đã định hình sự hiểu biết của tôi về khoa học dữ liệu như thế nào?**
- Dự án này đã giúp tôi nhận ra rằng khoa học dữ liệu không chỉ là việc lập mô hình — mà là một quy trình làm việc hoàn chỉnh:

**chất lượng dữ liệu → làm sạch → khám phá → trực quan hóa cẩn thận → phân tích mạnh mẽ → diễn giải thực tiễn.**
- Nó cũng củng cố tầm quan trọng của việc viết các phân tích có thể hỗ trợ các quyết định thực tế:

- Câu hỏi 3 yêu cầu đưa ra khuyến nghị trong điều kiện ràng buộc (số lượng cảm biến hạn chế),

- Câu hỏi 4 yêu cầu lập luận dựa trên bằng chứng về một cơ chế thực tế (gió và chất lượng không khí).
